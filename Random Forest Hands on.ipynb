```python
# label encoding - to convert object data into numerical

col_list = [] #to add all the object type columns into the list

for col in df.columns:      # col is my iterator
    if(df[col].dtype == "object"):
        col_list.append(col)
    print(col_list)
```

```python
col_list.pop(-1)
```

```python
print(col_list)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
for i in col_list:
    df[i] = label_encoder.fit_transform(df[i])
# fit ---> giving the data, transform --> converting the data into numerical form
```

```python
df
```

```python
df.dtypes
```

```python
# Dividing the data into x and y - Independent and Dependent Feature

x = df.iloc[:,:-1] # features or Independent variables
y = df.iloc[:,-1] # target data
```

```python
x
```

```python
y
```

```python
# Splitting the data to train and test
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 44)
```

```python
# Building the Random Forest Model
classifier1 = RandomForestClassifier(n_estimators=400) # No. of decision trees = 400
classifier1.fit(x_train, y_train) # Training the model
```

```python
y_pred = classifier1.predict(x_test) # predictions for testing data
```

```python
y_pred
```

```python
y_test
```

```python
a1 = accuracy_score(y_test,y_pred)
a1
```

```python
n = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
accuracy_list = []

for i in n:
        Rd_model = RandomForestClassifier(n_estimators=i)
        Rd_model.fit(x_train, y_train)
        pred = Rd_model.predict(x_test)
        a = accuracy_score(y_test, pred)
        accuracy_list.append(a)
accuracy_list
```

```python
d = {"Numbers of Trees":n, "Accuracy": accuracy_list}
df1 = pd.DataFrame(d)
print(df1)
```
